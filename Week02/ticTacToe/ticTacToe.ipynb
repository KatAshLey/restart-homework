```python
#displays a tic tac toe grid using " " | - +

#parts of board
s = " "
v = "|"
h = "-"
j = "+"

#variables set up
size = 1
space_line_base = [s, v, s, v, s]
dash_line_base = [h, j, h, j, h]
space_line = []
dash_line = []
board_base = [space_line, dash_line, space_line, dash_line, space_line]
board = []

#ask for user input for size of tictactoe board
size = int(input("What size board for tic tac toe?"))

if size <= 0:
    size = int(input("Needs to be bigger. What size board for tic tac toe?"))
elif size >= 11:
    size = int(input("Woah!!! Too big, try smaller. What size board for tic tac toe?"))

col_size = size*3 + 2
row_size = size*3 + 2

#extends space for size
for sL in space_line_base:
    if sL == s:
        space_line.extend(sL * size)
    else:
        space_line.extend(sL*1)
        
#extends dash for size
for dL in dash_line_base:
    if dL == h:
        dash_line.extend(dL * size)
    else:
        dash_line.extend(dL*1)

#extend height for size
for line in board_base:
    if line == space_line:
        board.extend(line * size)
    else:
        board.extend(line * 1)

#breaks 1d array into 2d array
length = len(board)
size_board = [board[i*length // col_size: (i+1)*length // col_size]
            for i in range(col_size)]


""" print()
print(size)
print(colSize)
print(spaceLine)
print(dashLine)
print(boardBase)

#print(board)
#print(sizeBoard)
print()"""


#prints board if its a 2d array
for i in range(len(size_board)):
    for j in range(len(size_board[i])):
        print(size_board[i][j], end="")
    print()
    
    
#convert list to string to be able to write to text file
str_board_lines = ["".join(line) for line in size_board] #for better spacing change "" to " "
#print(strBoardLines)
str_board_text = ["\n".join(str_board_lines)]
#print(strBoardText)

#write to text file
with open("ticTacToe.txt", "w") as f:
    f.writelines(str_board_text)
```
